#**Measures of Frequency**

###**Q1: Load complaints.csv dataset**

```python
import pandas as pd
```

```python
complaints = pd.read_csv('/content/complaints (1).csv')
complaints
```

```python
frequency = complaints['Issue'].value_counts()
```

### **Q2: Group the complaints by issue type**

> Hint: use df.groupby()



```python
complaints_by_issue = complaints.groupby('Issue').size().reset_index(name='frequency')
complaints_by_issue
```

### **Q3: Calculate the relative frequency**

```python
relative_frequency = frequency / len(complaints)
relative_frequency
```

```python
complaints_by_issue['relative_frequency'] = complaints_by_issue['frequency'] / len(complaints)
complaints_by_issue
```

### **Q4: Calculate the cumulative frequency**

> Hint: use column_name.cumsum()



```python
cumulative_frequency = frequency.cumsum()
cumulative_frequency
```

```python
complaints_by_issue['cumulative_frequency'] = complaints_by_issue['frequency'].cumsum()
complaints_by_issue
```

### **Q5: Calculate the cumulative relative frequency**

```python
cumulative_frequency_relative = cumulative_frequency / len(complaints)
cumulative_frequency_relative
```

```python
complaints_by_issue['cumulative_relative_frequency'] = complaints_by_issue['cumulative_frequency'] / len(complaints)
complaints_by_issue
```

### **Q6: Calculate the quartiles**

> Hint: use columnName.quantile(), pd.cut()



```python
quartiles = frequency.quantile([0.25, 0.5, 0.75])
quartiles
```

```python
complaints_quartile = pd.cut(frequency,bins=3,labels=(['Q1', 'Q2', 'Q3']))
```

```python
complaints_quartile
```

```python
quartiles = complaints_by_issue['frequency'].quantile([0.25, 0.5, 0.75])
complaints_by_issue['quartile'] = pd.cut(complaints_by_issue['frequency'], bins=3, labels=(['Q1', 'Q2', 'Q3']))
complaints_by_issue
```

### **Q7: Calculate the percentiles**

> Hint: use columnName.quantile(), pd.cut()


```python
percentiles = complaints_by_issue['frequency'].quantile([0.1, 0.25, 0.5, 0.75, 0.9])
complaints_by_issue['percentile'] = pd.cut(complaints_by_issue['frequency'], bins=5, labels=['P10', 'P25', 'P50', 'P75', 'P90'])
complaints_by_issue
```

# **Measures of Central Tendency**

### **Q1: Load titanic.csv dataset**

```python
titanic_df = pd.read_csv("/content/titanic (1).csv")
titanic_df.dropna(inplace=True)
```

### **Q2: Calculate the mean, median, and mode of the ages**

```python

ages = titanic_df['Age']
mean_age = np.mean(ages)
median_age = np.median(ages)
mode_age = ages.mode()[0]

print("Mean Age:", mean_age)
print("Median Age:", median_age)
print("Mode Age:", mode_age)
```

###**Q3: Calculate the mean, median, and mode of the fares by passenger class**

```python
titanic_df['Fare'].value_counts().index[0]
```

```python
titanic_df.groupby('Pclass')['Fare'].groups
```

```python
titanic_df.head()
```

```python
fares_by_class = titanic_df.groupby('Pclass')['Fare'].agg(['mean', 'median', lambda x:x.value_counts().index[0]])

# Rename the columns for clarity
fares_by_class = fares_by_class.rename(columns={'mean': 'Mean Fare', 'median': 'Median Fare', '<lambda_0>': 'Mode Fare'})

print(fares_by_class)
```

# **Measures of Dispersion**

### **Q1: Read in the sales_data.csv dataset**

```python
sales_data = pd.read_csv('sales_data.csv')
```

### **Q2: Calculate the range**

```python
def range_fun(col):
  return col.max() - col.min()
```

```python
print('Order_Quantity Range: {:.2f}'.format(range_fun(sales_data['Order_Quantity'])))
print('Unit_Cost Range: {:.2f}'.format(range_fun(sales_data['Unit_Cost'])))
print('Unit_Price Range: {:.2f}'.format(range_fun(sales_data['Unit_Price'])))
print('Profit Range: {:.2f}'.format(range_fun(sales_data['Profit'])))
print('Cost Range: {:.2f}'.format(range_fun(sales_data['Cost'])))
print('Revenue Range: {:.2f}'.format(range_fun(sales_data['Revenue'])))
```

### **Q3: Calculate the variance and standard deviation**

> Hint: use np.var() and np.std()



```python
def var_std(col):
  return np.var(col), np.std(col)
```

```python
print('Order_Quantity Variance and Standard Deviation:',(var_std(sales_data['Order_Quantity'])))

print('Unit_Cost Varianceand Standard Deviation:', (var_std(sales_data['Unit_Cost'])))

print('Unit_Price Variance and Standard Deviation:',(var_std(sales_data['Unit_Price'])))

print('Profit Variance and Standard Deviation:',(var_std(sales_data['Profit'])))

print('Cost Variance and Standard Deviation:',(var_std(sales_data['Cost'])))

print('Revenue Variance and Standard Deviation:',(var_std(sales_data['Revenue'])))
```

### **Q4: Calculate the coefficient of variation**

```python
def coff_variation(std, col):
   return std/col.mean() *100
```

```python
_, Order_Quantity_std = var_std(sales_data['Order_Quantity'])
print('Order Quantity Coefficient of Variation: {:.2f}%'.format(coff_variation(Order_Quantity_std, sales_data['Order_Quantity'])))
```

```python
_, Unit_Cost_std = var_std(sales_data['Unit_Cost'])
print('Unit Cost Coefficient of Variation: {:.2f}%'.format(coff_variation(Order_Quantity_std, sales_data['Unit_Cost'])))
```

```python
_, Unit_Price_std = var_std(sales_data['Unit_Price'])
print('Unit_Price Coefficient of Variation: {:.2f}%'.format(coff_variation(Order_Quantity_std, sales_data['Unit_Price'])))
```

```python
_, Profit_std = var_std(sales_data['Profit'])
print('Profit Coefficient of Variation: {:.2f}%'.format(coff_variation(Order_Quantity_std, sales_data['Profit'])))
```

```python
_, Cost_std = var_std(sales_data['Cost'])
print('Cost Coefficient of Variation: {:.2f}%'.format(coff_variation(Order_Quantity_std, sales_data['Cost'])))
```

```python
_, Revenue_std = var_std(sales_data['Revenue'])
print('Revenue Coefficient of Variation: {:.2f}%'.format(coff_variation(Revenue_std, sales_data['Revenue'])))
```

### **Q5: Calculate the interquartile range**

> Hint: use .quantile()



```python
def iqr_sales(col):
  return col.quantile(0.75) - col.quantile(0.25)
```

```python
print('Order_Quantity Interquartile Range: {:.2f}'.format(iqr_sales(sales_data['Order_Quantity'])))

print('Unit_Cost Interquartile Range: {:.2f}'.format(iqr_sales(sales_data['Unit_Cost'])))

print('Unit_Price Interquartile Range: {:.2f}'.format(iqr_sales(sales_data['Unit_Price'])))

print('Profit Interquartile Range: {:.2f}'.format(iqr_sales(sales_data['Profit'])))

print('Cost Interquartile Range: {:.2f}'.format(iqr_sales(sales_data['Cost'])))

print('Revenue Interquartile Range: {:.2f}'.format(iqr_sales(sales_data['Revenue'])))
```
