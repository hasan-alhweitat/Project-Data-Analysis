1. <b>For categorical data, simple proportions or percentages tell the story of the data. </b> </br>
2. For visualization <b> bar charts </b> show the frequency or proportion for each category plotted as bars.
3. <b> Bar charts </b>, seen often in the popular press, are a common visual tool for displaying a single categorical variable. Categories are listed on the x-axis, and frequencies or proportions on the y-axis
4. Note that a <b> bar chart resembles a histogram </b>; in a bar chart the x-axis represents different categories of a factor variable, while in a histogram the x-axis represents values of a single variable on a numeric scale. 
5. In a histogram, the bars are typically shown touching each other, with gaps indicating values that did not occur in the data. In a bar chart, the bars are shown separate from one another.
