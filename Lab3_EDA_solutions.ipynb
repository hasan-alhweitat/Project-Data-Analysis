Types of analysis for Data Visualization:
1. <b> Univariate Analysis </b>: In the univariate analysis, we will be using a single feature to analyze almost all of its properties. 

2. <b> Bivariate Analysis</b>: When we compare the data between exactly 2 features then it is known as bivariate analysis. 

3. <b> Multivariate Analysis</b>: In the multivariate analysis, we will be comparing more than 2 variables.

# **Univariate Analysis**

### Bar chart

```python
df['Survived'].value_counts()
```

```python
sns.countplot(x="Survived", data=df)
plt.show()
```

```python
#hue= adds an additional variable to plot via a color mapping
sns.countplot(x="Survived",hue="Sex", data=df)
plt.show()
```

```python
sns.countplot(x="Pclass", data=df)
plt.show()
```

### Distribution plot

```python
sns.displot(x="Age", data=df) 
plt.show()
```

### Condition the distribution plot

```python
df['Age'].describe()
```

```python
#hue= adds an additional variable to plot via a color mapping
sns.displot(x="Age", hue='Sex', data=df)
plt.show()
```

```python
#let us check the 5 number summary of the Fare column
df['Fare'].describe()
```

```python
sns.displot(x="Fare", hue='Survived', data=df, height=8, aspect=1.5)
plt.show() #doc sns
```

### Box plot

```python
df['Age'].describe()
```

```python
sns.boxplot(x="Age",y ="Sex",data=df)
plt.show()
```

```python
sns.boxplot(y='Age',x='Sex', data=df)
plt.show()
```

```python
sns.boxplot(x="Age", y="Sex", data=df)
```

```python
sns.boxplot(y='Fare',x='Sex', data=df)
plt.show()
```

```python
sns.boxplot(y='Age',x='Survived', data=df)
plt.show()
```

```python
sns.boxplot(y='Fare',x='Survived', data=df)
plt.show()
```

### Violin plot

```python
sns.violinplot(y='Age',data=df)
plt.show()
```

# **Bivariate analysis**

### Line plot

```python
sns.lineplot(x="Age", y="Fare", data=df)
plt.show()
```

### Scatter plot

```python
sns.scatterplot(x="Age", y="Fare", data=df)
plt.show()
```

### Pair plot

```python
sns.pairplot(df)
plt.show()
```

### Categorical data visualization

```python
sns.catplot(x="Sex", y="Fare", hue="Survived", kind="bar", data=df)
plt.show()
```

```python
df.groupby(["Sex","Survived"]).Fare.mean()
```
